# Ploting paper graphs

```python
import pandas as pd
import seaborn as sns
import matplotlib as plt
```

```python
result_path = "/home/joel/resultsAsoc2021/outros_experimentos_testes/MOA1_noUpdateMtxMatrizByClassification/"
```

```python
data_minasbr = pd.read_csv(result_path+"/MINAS-BR-Results.csv")
data_minasbr
```

```python
concept_evolution_window = pd.read_csv(result_path+"conceptEvolution-info.csv")
concept_evolution_window
```

```python
np_windows = pd.read_csv(result_path+"NP-info.csv")
np_windows = np_windows.dropna().reset_index().drop(['index'],axis=1)
np_windows.head()
```

```python
x = np_windows[' windowAssociation '].duplicated().map(lambda i: not i)
np_windows = np_windows[x]
np_windows.reset_index(inplace=True)
np_windows.drop(['index'],axis=1,inplace=True)
np_windows.sort_values(by=np_windows.columns[-1],inplace=True)
```

```python
ax = sns.lineplot(data=data_minasbr, y="F1M", x=data_minasbr.index)
sns.lineplot(ax=ax,data=data_minasbr,y="unkRM",x=data_minasbr.index)
ax.set_ylim(0,1)
ax.figure.set_size_inches(12,6)

for i in concept_evolution_window.index:
    ax.annotate(concept_evolution_window['label'].iloc[i],
                xy=(concept_evolution_window['window'].iloc[i]-1, 0), 
                xycoords='data',
                bbox=dict(boxstyle="round", fc="red", ec="red"),
                xytext=(0, -40),
                textcoords='offset points',
                ha='center',
                arrowprops=dict(arrowstyle="->"),
                fontsize=16)
pos = [60,40,20,40,20,20,20,40,20]
for i in np_windows.index:
    label = str(np_windows[' AssociatedClass'].iloc[i])
    ax.annotate(label, xy=(np_windows.iloc[i,-1]-1, 1),  xycoords='data',
                xytext=(0, pos[i]), textcoords='offset points', ha='center',
                arrowprops=dict(arrowstyle="->"),
                bbox=dict(boxstyle="round", fc="cyan", ec="white"),
                fontsize=12)

    ax.vlines(x=np_windows.iloc[i,-1]-1,
            ymin=0,
            ymax=1,
            colors='black',
            linestyles='solid',
            label="NP1")

ax.set_xlabel("Windows", fontsize=18)
ax.set_ylabel("F1M and UnkRM", fontsize=18)
ax.tick_params(labelsize=12)
```
